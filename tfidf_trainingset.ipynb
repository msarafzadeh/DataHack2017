```python
import pandas as pd
import string
import nltk
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

### import data 

```python
#extract from train data 
```

```python
df_train = pd.read_csv("NEDDataHack2017_train.tsv", header=0, sep = "\t")
```

```python
textTrain = df_train['text'].str.lower().to_string(index=False)
```

```python
textTrain = str.replace(textTrain, '<p>', " ");
```

```python
translator = str.maketrans('', '', string.punctuation)
textTrain = textTrain.translate(translator)
```

```python
textTrain = textTrain.split('\n')
```

```python
textTrain = [x.strip(' \t') for x in textTrain]
```

```python
#extract from abstract additional data
```

```python
df_adtl = pd.read_csv("test.csv", header=0)
```

```python
textAdtl = df_adtl['Abstract'].to_string(index=False)
```

```python
textAdtl = str.replace(textAdtl, '<p>', " ");
```

```python
translator = str.maketrans('', '', string.punctuation)
textAdtl = textAdtl.translate(translator)
```

```python
textAdtl = textAdtl.split('\n')
```

```python
textAdtl = [x.strip(' \t') for x in textAdtl]
```
